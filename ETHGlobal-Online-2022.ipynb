```python
from datetime import datetime
```

```python
from subgrounds import Subgrounds
```

```python
sg=Subgrounds()
```

```python
# load unlock subgraph 
unlock = sg.load_subgraph('https://api.thegraph.com/subgraphs/name/unlock-protocol/unlock')
```

```python
from subgrounds.subgraph import SyntheticField

# registrationdate synthetic field
unlock.Key.createdDate = SyntheticField(
    lambda timestamp: str(datetime.fromtimestamp(timestamp)),
    SyntheticField.STRING,
    unlock.Key.createdAt
)

unlock.Lock.priceEth = SyntheticField(
    lambda price: price / (10**18),
    SyntheticField.FLOAT,
    unlock.Lock.price
)

```

```python
keys = unlock.Query.keys(
    first=1000, # latest 1000 registrations
    orderBy=unlock.Key.createdAt, 
    orderDirection="desc" 
)

field_paths = [
    keys.lock.name,
    keys.owner.address,
    keys.createdDate,
    keys.lock.priceEth,
    keys.expiration
]
```

```python
# get data
df = sg.query_df(field_paths)

# print the first five results
df.head(100)
```
